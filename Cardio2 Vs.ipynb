```python
import pandas as pd

# Load the dataset
data = pd.read_csv(r"C:\Users\amkb2\OneDrive\Desktop\DoctorAI\cardio_train.csv", delimiter=";")
```

```python
data.head()
```

```python
# Drop rows with missing values
data = data.dropna()
```

```python
import matplotlib.pyplot as plt

# Create a histogram of age
plt.hist(data['age'], bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Split the data into features and target variable
X = data.drop('cardio', axis=1)
y = data['cardio']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)

# Create and train a logistic regression model
model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

**Gaussian ,GradienBoost,ADABoost,XGBoost**

```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import sklearn
```

```python
data_vas =  pd.read_csv(r"C:\Users\amkb2\OneDrive\Desktop\DoctorAI\cardio_train.csv", delimiter=";")
```

```python
data_vas.head()
```

```python
# Drop rows with missing values
data_vas = data_vas.dropna()
```

```python
import matplotlib.pyplot as plt

# Create a histogram of age
plt.hist(data_vas['age'], bins=30)
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Age')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Split the data into features and target variable
X = data_vas.drop('cardio', axis=1)
y = data_vas['cardio']
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

**Gaussian**

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(X_train, y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

```python
# Evaluate the model's accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
from sklearn.metrics import  confusion_matrix
cm_NBClassifier = confusion_matrix(y_test,y_pred)
```

```python
print("confusion_matrix : ",cm_NBClassifier)
```

```python
from sklearn.ensemble import (RandomForestClassifier,
                              AdaBoostClassifier,
                              GradientBoostingClassifier)
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
rf =RandomForestClassifier(n_estimators=500)
rf.fit(X_train,y_train)
y_pred = rf.predict(X_train)
accuracy_score(y_train,y_pred)
```

```python
y_pred = rf.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import  confusion_matrix
cm_Randomforest = confusion_matrix(y_test,y_pred)
print("Confusion metrics: ",cm_Randomforest)
```

**AdaBoost**

```python
ada =AdaBoostClassifier(n_estimators=500)
ada.fit(X_train,y_train)
y_pred = ada.predict(X_train)
accuracy_score(y_train,y_pred)
```

```python
y_pred = ada.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import  confusion_matrix
cm_AdaBoost = confusion_matrix(y_test,y_pred)
print("Confusion metrics: ",cm_AdaBoost)
```

**Gradient Boosted Tree**

```python
gb =GradientBoostingClassifier(n_estimators=500)
gb.fit(X_train,y_train)
y_pred = gb.predict(X_train)
accuracy_score(y_train,y_pred)
```

```python
y_pred = gb.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import  confusion_matrix
cm_Gradient = confusion_matrix(y_test,y_pred)
print("Confusion metrics: ",cm_Gradient)
```

**XGBoost**

```python
xgb =XGBClassifier(n_estimators=500)
xgb.fit(X_train,y_train)
y_pred = xgb.predict(X_train)
accuracy_score(y_train,y_pred)
```

```python
y_pred = xgb.predict(X_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.metrics import  confusion_matrix
cm_XGB = confusion_matrix(y_test,y_pred)
print("Confusion metrics: ",cm_XGB)
```
